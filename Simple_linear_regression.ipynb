Linear Regression Problem

Predicting the price for a house using the size features

```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/real_estate_price_size.csv')
```

```python
df.head()
```

```python
df.shape
```

Visualizing the relationship

```python
#using scatter plot
plt.figure(figsize=(8,4))
plt.scatter(x='size', y='price', data=df)
plt.xlabel('size')
plt.ylabel('price')
plt.title('Relationship between size and price')
plt.show()
```

```python
#split into features and target variable
x = df.drop(columns='price').values
y = df.drop(columns='size').values
```

Fitting in the best line

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#train model on dataset
model.fit(x,y)
```

```python
#to make prediction
pred = model.predict(x)
pred[0]
```

```python
y[0]
```

```python
#using scatter plot
plt.figure(figsize=(8,4))
plt.scatter(x='size', y='price', data=df)
plt.plot(x, pred, color='red')
plt.xlabel('size')
plt.ylabel('price')
plt.title('Relationship between size and price')
plt.show()
```
