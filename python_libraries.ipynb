```python
#sample DataFrame with missing values
data = {
    'A': [1, 2, 3, 4, 5, 6, 7],
    'B': [5, np.nan, np.nan, 8, 11, 15, np.nan],
    'C': [10, 11, 12, np.nan, np.nan, 4, 6]
}
df = pd.DataFrame(data)
```

```python
df.isna().sum() #missing values func
```

```python
#dealing with missing values
#fill with 0, mean, median and mode
#drop missing values
#forward or backward fill
```

```python
X = df.dropna() # parameter axis is to check what to drop 0-rows and 1-columns by default, it is 0
```

```python
X #the rows with missing values has been dropped
```

```python
Y = df.dropna(axis=1) #specify that missing columns should be dropped
```

```python
Y #missing columns has been dropped
```

```python
#fill with 0
Z = df.fillna(0) #0 will replace missing values
```

```python
Z
```

```python
#fill with mean
C = df.fillna(df.mean()) #check each column and find the mean
```

```python
C
```

```python
#forward or backward fill
bf = df.fillna(method='bfill')
#fill nan/missing value with the following value behind it.
```

```python
bf #the last missing value unfilled because it has no following value
```

```python
ff = df.fillna(method='ffill') #fill missing value with proceeding value
```

```python
ff
```

```python
#new method of doing ffill and bfill
nmf = df.ffill() #for forward fill
```

```python
nmf
```

```python
nmb = df.bfill() #new way of doing backward fill
```

```python
nmb
```

```python
raw_data_1 = {
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'role': ['Python Developer', 'Python Developer', 'PHP Developer', 'C++ Developer', 'Data '],
        'experience': [3,4,4,3,2],
        'qualification': ['Graduate', 'Graduate', 'Post Graduate', 'Graduate', 'Post Graduate']
}

raw_data_2 = {
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce'],
        'role': ['Data Analyst', 'ML Engineer', 'Data Scientist', 'Business Analyst'],
        'experience': [2, 6, 4, 7],
        'salary': ['30k', '45k', '60k', '35k'],
        'awards_collected': [0, 2, 3, 1]
}
```

```python
df1 = pd.DataFrame(raw_data_1)
df2 = pd.DataFrame(raw_data_2)
```

```python
#combing data
#datasets can be concatenated in 2 ways;
#row axis and column axis #remember 0=row axis
#1= column axis
pd.concat([df1, df2], axis=0) #combine rows
```

```python
#in the previous output, the index is not in the right form, to correct that.
pd.concat([df1, df1], axis=0, ignore_index=True) #the contact func has an ignore_index parameter
#the index has been corrected and in right order
```

```python
pd.concat([df1, df2], axis=1) #but this is not a good dataset
```

Merge
-inner join
-left outer join
-right outer join
-full outer join
the merge function takes left and right parameters which are datasets to be merged







```python
raw_data_1 = {
        'id': ['A1', 'A2', 'A3', 'A4', 'A5'],
        'project_id_initial': ['DX13', 'DX11','DX29', 'DX30', 'DX28'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'role': ['Python Developer', 'Python Developer', 'PHP Developer', 'C++ Developer', 'Data '],
        'experience': [3,4,4,3,2],
        'qualification': ['Graduate', 'Graduate', 'Post Graduate', 'Graduate', 'Post Graduate']
}

raw_data_2 = {
        'id': ['A1', 'A2', 'A3', 'A6'],
        'project_id_final': ['DX13', 'DX11', 'DX30', 'DX29'],
        'salary': ['30k', '45k', '60k', '35k'],
        'awards_collected': [0, 2, 3, 1]
}
```

```python
df1 = pd.DataFrame(raw_data_1)
df2 = pd.DataFrame(raw_data_2)
```

```python
pd.merge(df1,df2) #how parameter by default is inner merge
```

```python
#if you specify on, it merges on the particular column inputted, but by default, the ID is the only common default
pd.merge(df1, df2, how='left') #concentrate on left data
```

```python
pd.merge(df1, df2, how='right') #focus on right data
```

```python
pd.merge(df1, df2, how='outer')#merge all
```

```python
data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Location': ['New York', 'Los Angeles', 'New York', 'Los Angeles', 'New York', 'Los Angeles'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [50000, 60000, 70000, 80000, 90000, 100000],
    'Bonus': [5000, 6000, 7000, 8000, 9000, 10000]
}
df = pd.DataFrame(data)
```

Slicing- selecting a number of rows and columns
*   loc: labeled based indexing
*   iloc: integer-based indexing




```python
#loc- have to specify column names
df.loc[:, ['Department', 'Employee', 'Salary']]
```

```python
#assuming you want it from index 1-4
df.loc[1:4, ['Department', 'Employee', 'Salary']]
```

```python
#iloc- you have to know index num of column
df.iloc[:,[0, 2,3]]
```
