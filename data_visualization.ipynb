```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creating a simple plot
#line plot- majory for trend; seeing growth of a thing over a period of tine.
apples_yield = [0.895, 0.91, 0.919, 0.926, 0.931, 0.928]
plt.plot(apples_yield);
```

```python
apples_yield
years = [2010,2011,2012,2013,2014,2015]
plt.plot(years,apples_yield); #x & y axia
```

```python
#labeling axis and giving title
plt.plot(years,apples_yield);
plt.xlabel('Years')
plt.ylabel('Apple Yield')
plt.title('Yearly Yield of Apple')
plt.show()
```

Plotting multiple lines

```python
years = range(2010,2016)
years
```

```python
years
apples_yield = [0.895,0.91,0.919,0.926,0.931,0.928]
oranges_yield = [0.99,0.978,0.985,0.908,0.889,0.878]
```

```python
plt.plot(years,apples_yield)
plt.plot(years,oranges_yield)
plt.xlabel('Years')
plt.ylabel('Yields (Tons)')
plt.title('Yearly Yield Rate of Apples and Oranges')
plt.show()
```

```python
#in the above output, there is no way to signify which plot is for apple or prange
plt.plot(years, apples_yield)
plt.plot(years, oranges_yield)
plt.xlabel('Years')
plt.ylabel('Yields (Tons)')
plt.title('Yearly Yield Rate of Apples and Oranges')
plt.legend(['Apple','Orange'])
plt.show()
```

```python
#marker for point identification
plt.plot(years, apples_yield, marker= 'o')
plt.plot(years, oranges_yield, marker= '*')
plt.xlabel('Years')
plt.ylabel('Yields (Tons)')
plt.title('Yearly Yield Rate of Apples and Oranges')
plt.legend(['Apple','Orange'])
plt.show()
```
